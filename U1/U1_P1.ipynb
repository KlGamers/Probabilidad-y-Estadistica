```python
from statistics import *

ga = [70, 80, 84, 90, 67, 87, 90, 98, 78, 76, 96]
print("La lista de calificaciones del primer grupo es: " + str(ga))
print("La media de la lista del grupo es: ", sum(ga)/len(ga))
print("La mediana de la lista del grupo es: ", median(ga))
print("La moda de la lista del grupo es: ", mode(ga))
print("La varianza de la lista del grupo es: ", variance(ga))
print("El rango de la lista del grupo es: ", max(ga) - min(ga))
print("Los cuartiles de la lista del grupo son: ", quantiles(ga, n=4))
print("La desviación estandar de la lista del grupo es: ", stdev(ga))
```

```python
import random

f1=[]
for i in range(100):
  na = round(random.uniform(50, 100), 2)
  f1.append(na)
print(' '.join(map(str, f1)))
```

```python
import random

f2=[]
for i in range(100):
  na = round(random.uniform(70, 100), 2)
  f2.append(na)
print(' '.join(map(str, f2)))
```

```python
import matplotlib.pyplot as plt

plt.hist(f1)
plt.title("Producción diaria de la fábrica 1: ")

print("La media es: ", sum(f1)/len(f1))

print("La mediana es: ", median(f1))

print("La moda es: ", mode(f1))

print("La varianza es: ", variance(f1))

print("El rango es: ", max(f1) - min(f1))

print("Los cuartiles son: ", quantiles(f1, n=4))

print("La desviación estandar es: ", stdev(f1))
```
