```python
import random
import matplotlib.pyplot as ptl
from statistics import *

def generate_production_data(num_samples, min_val, max_val):
  data = []
  for _ in range(num_samples):
    data.append(round(random.uniform(min_val, max_val), 2))
  return data

def plot_factory_boxplots(data1, title1, data2, title2, fig_title=None, y_lim=None):
  fig, (ax1, ax2) = ptl.subplots(nrows=1, ncols=2, figsize=(10, 5))

  bplot1 = ax1.boxplot(data1, vert=True, patch_artist=True)
  for patch in bplot1['boxes']:
    patch.set_facecolor('skyblue')
  ax1.set_title(title1, fontsize=14)
  ax1.set_ylabel('Producción en mts.', fontsize=12)
  ax1.tick_params(axis='y', labelsize=10)
  if y_lim:
    ax1.set_ylim(y_lim[0], y_lim[1])
  ax1.grid(axis='y', linestyle='--', alpha=0.7)

  bplot2 = ax2.boxplot(data2, vert=True, patch_artist=True)
  for patch in bplot2['boxes']:
    patch.set_facecolor('lightcoral')
  ax2.set_title(title2, fontsize=14)
  ax2.set_ylabel('Producción en mts.', fontsize=12)
  ax2.tick_params(axis='y', labelsize=10)
  if y_lim:
    ax2.set_ylim(y_lim[0], y_lim[1])
  ax2.grid(axis='y', linestyle='--', alpha=0.7)

  if fig_title:
      fig.suptitle(fig_title, fontsize=16, y=1.02)

  ptl.tight_layout(rect=[0, 0.03, 1, 0.95])
  ptl.show()


fabrica1 = generate_production_data(100, 50, 100)
fabrica2 = generate_production_data(100, 40, 93)

print("Primeros 10 datos de Fábrica 1:", fabrica1[:10])
print("Primeros 10 datos de Fábrica 2:", fabrica2[:10])

plot_factory_boxplots(fabrica1, 'Producción de la Fábrica 1',
                      fabrica2, 'Producción de la Fábrica 2',
                      fig_title='Comparación Inicial de Producción de Fábricas')

fabrica1.append(150)
print("\nÚltimo dato de Fábrica 1 después de añadir un valor atípico:", fabrica1[-1:])

plot_factory_boxplots(fabrica1, 'Fábrica 1 (con Atípico)',
                      fabrica2, 'Fábrica 2',
                      fig_title='Comparación de Producción (Fábrica 1 con Valor Atípico)')

plot_factory_boxplots(fabrica1, 'Fábrica 1 (con Atípico)',
                      fabrica2, 'Fábrica 2',
                      fig_title='Comparación de Producción (Límites Ajustados)',
                      y_lim=(30, 160))
```
