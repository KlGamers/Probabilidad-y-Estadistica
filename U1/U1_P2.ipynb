```python
import random
import matplotlib.pyplot as plt
from statistics import *

f1 = []
for i in range(100):
    na = round(random.uniform(50, 100),2)
    f1.append(na)
print("Fabrica 1:", f1)

f2 = []
for i in range(100):
    na = round(random.uniform(50, 100),2)
    f2.append(na)
print("Fabrica 2:", f2)

m1 = max(f1)
m2 = max(f2)

min1 = min(f1)
min2 = min(f2)

r1 = m1 - min1
r2 = m2 - min2

me1 = mean(f1)
me2 = mean(f2)

med1 = median(f1)
med2 = median(f2)

mod1 = mode(f1)
mod2 = mode(f2)

print("Estadistica descriptiva de la fabrica 1 \n", "Media: ", me1, "\n Mediana: ", med1, "\n Moda: ", mod1, "\n Máxmo: ", m1, "\n Mínimo: ", min1, "\n Rango: ", r1)
print()
print("Estadistica descriptiva de la fabrica 2 \n", "Media: ", me2, "\n Mediana: ", med2, "\n Moda: ", mod2, "\n Máxmo: ", m2, "\n Mínimo: ", min2, "\n Rango: ", r2)
```
