```python
import random
import matplotlib.pyplot as plt
from statistics import *
import seaborn as sns

sns.set_style("whitegrid")

f1 = []
for i in range(100):
    na = round(random.uniform(50, 100),2)
    f1.append(na)
print("Fabrica 1:", f1)

f2 = []
for i in range(100):
    na = round(random.uniform(50, 100),2)
    f2.append(na)
print("Fabrica 2:", f2)
print()

m1 = max(f1)
m2 = max(f2)

min1 = min(f1)
min2 = min(f2)

r1 = m1 - min1
r2 = m2 - min2

me1 = mean(f1)
me2 = mean(f2)

med1 = median(f1)
med2 = median(f2)

mod1 = mode(f1)
mod2 = mode(f2)

std1 = stdev(f1)
std2 = stdev(f2)

print("Estadistica descriptiva de la fabrica 1 \n", "Media: ", me1, "\n Mediana: ", med1, "\n Moda: ", mod1, "\n Máxmo: ", m1, "\n Mínimo: ", min1, "\n Rango: ", r1, "\n Desviación Estándar: ", std1)
print()
print("Estadistica descriptiva de la fabrica 2 \n", "Media: ", me2, "\n Mediana: ", med2, "\n Moda: ", mod2, "\n Máxmo: ", m2, "\n Mínimo: ", min2, "\n Rango: ", r2, "\n Desviación Estándar: ", std2)

plt.figure(figsize = (8,5))
plt.hist(f1, bins=18, alpha=0.7, color='b', edgecolor= "black")
plt.axvline(me1, color='b', linestyle='dashed', linewidth=2, label=f'Media={me1:.2f}')
plt.axvline(med1, color='r', linestyle='dashed', linewidth=2, label=f'Mediana={med1:.2f}')
plt.axvline(mod1, color='y', linestyle='dashed', linewidth=2, label=f'Moda={mod1:.2f}')
plt.axvline(me1 - std1, color='black', linestyle='dashed', linewidth=2, label=f'Desv. Estándar={std1:.2f}')
plt.axvline(me1 + std1, color='black', linestyle='dashed', linewidth=2)
plt.xlabel('Valor de Producción')
plt.ylabel('Frecuencia')
plt.title('Distribución de la Producción Diaria para la Fábrica 1')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

print()
plt.figure(figsize = (8,5))
plt.hist(f2, bins=18, alpha=0.7, color='g', edgecolor= "black")
plt.axvline(me2, color='g', linestyle='dashed', linewidth=2, label=f'Media={me2:.2f}')
plt.axvline(med2, color='orange', linestyle='dashed', linewidth=2, label=f'Mediana={med2:.2f}')
plt.axvline(mod2, color='purple', linestyle='dashed', linewidth=2, label=f'Moda={mod2:.2f}')
plt.axvline(me2 - std2, color='black', linestyle='dashed', linewidth=2, label=f'Desv. Estándar={std2:.2f}')
plt.axvline(me2 + std2, color='black', linestyle='dashed', linewidth=2)
plt.xlabel('Valor de Producción')
plt.ylabel('Frecuencia')
plt.title('Distribución de la Producción Diaria para la Fábrica 2')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
print()
sns.histplot(x= f1, kde = True, kde_kws = {'bw_adjust':0.5}, bins = 10)
```
